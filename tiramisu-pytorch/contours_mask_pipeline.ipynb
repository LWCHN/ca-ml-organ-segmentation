# Metrics

```python
import torch
from torch.nn import BCELoss

def dice_coef(input, target):
    smooth = 1.

    iflat = input.view(-1)
    tflat = target.view(-1)
    #iflat = input.round()  # use only for evaluation
    #tflat = target.round()  # use only for evaluation
    intersection = (iflat * tflat).sum()
    
    return (2. * intersection + smooth) / (iflat.sum() + tflat.sum() + smooth)
```

```python
class DC_loss2(torch.nn.modules.loss._Loss):

    def __init__(self, weight=None, size_average=None, reduce=None, reduction='elementwise_mean', pos_weight=None):
        super(DC_loss2, self).__init__(size_average, reduce, reduction)
        self.register_buffer('weight', weight)
        self.register_buffer('pos_weight', pos_weight)

    def forward(self, input, target):
        return dice_coef(input, target)
```

# Training

```python
import shutil

def save_checkpoint(model, optimizer, train_epoch_metrics, val_epoch_metrics, epoch, is_best, logpath):

    state = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'train_epoch_metrics': train_epoch_metrics,
        'val_epoch_metrics': val_epoch_metrics   # could be test_epoch_metrics in the old models
    }

    filename = "{}.pth.tar".format(logpath)
    torch.save(state, filename)
    if is_best:
        shutil.copyfile(filename, '{}.best.pth.tar'.format(logpath))
```

```python
from tqdm import tqdm_notebook as tqdm

def run_epoch(mode, loader, net, criterion, criterion2, optimizer, alpha=0.99):
    if mode == 'train':
        net.train()
    else:
        net.eval()
    total_loss = 0
    total_dice = 0
    total_bce = 0
    for i, data in tqdm(enumerate(loader), total=len(loader)):
        inputs = data['image']
        targets = data['masks']

        inputs, targets = inputs.to(device), targets.to(device)

        with torch.set_grad_enabled(mode == 'train'):
            outputs = net(inputs)
            d = criterion(outputs, targets)
            b = criterion2(outputs, targets)
            loss = 1 - torch.log(d) + b

            total_loss += loss.item()
            total_dice += d.item()
            total_bce += b.item()

            if mode == 'train':
                optimizer.zero_grad()

                loss.backward()
                optimizer.step()


    print('finished epoch', epoch, '{} mode'.format(mode), 'loss', total_loss / (i + 1))

    return {'loss': total_loss / (i + 1), 'dice': total_dice / (i + 1), 'bce': total_bce / (i + 1)}
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
criterion.to(device)
criterion2.to(device)

tiramisu.to(device)

train_epoch_metrics = {
        'loss': [],
        'dice': [],
        'bce': []
    }
val_epoch_metrics = {
        'loss': [],
        'dice': [],
        'bce': []
    }

is_best = False
best_loss = 0

for epoch in range(200):
    train_metrics = run_epoch('train', trainloader, tiramisu, criterion, criterion2, optimizer)
    valid_metrics = run_epoch('test', validloader, tiramisu, criterion, criterion2, optimizer)
    
    if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau):
        scheduler.step(valid_metrics['loss'])
    else:
        scheduler.step()
    
    for m in valid_metrics:
        train_epoch_metrics[m].append(train_metrics[m])
        val_epoch_metrics[m].append(valid_metrics[m])
    
    if epoch == 0 or valid_metrics['loss'] < best_loss:
        is_best = True
        best_loss = valid_metrics['loss']
        
    save_checkpoint(tiramisu, optimizer, train_epoch_metrics, 
                    val_epoch_metrics, epoch, is_best, './models/6_channels_fold1_input_minus_1_to_1')
    
    is_best = False
    

print('Finished Training')
```

```python
import numpy as np

def iou_bin(y_true, y_pred):
    ious = []
    for im1, im2 in zip(y_true, y_pred):
        im1 = np.round(np.asarray(im1.detach().cpu()))
        im2 = np.round(np.asarray(im2.detach().cpu()))
        im1 = im1.astype(np.bool)
        im2 = im2.astype(np.bool)
        intersection = np.logical_and(im1, im2)
        union = np.logical_or(im1, im2)
        iou = np.sum(intersection) / np.sum(union)
        ious.append(iou)
    return np.mean(ious)

def dice_coef_thresh(input, target):
    smooth = 1e-5

    iflat = input.view(-1)
    tflat = target.view(-1)
    iflat = input.round()  # use only for evaluation
    tflat = target.round()  # use only for evaluation
    intersection = (iflat * tflat).sum()
    
    return (2. * intersection + smooth) / (iflat.sum() + tflat.sum() + smooth)
```

```python
d = dice_coef_thresh(outputs, tars)
b = criterion2(outputs, tars)
iou = iou_bin(outputs, tars)
loss = 1 - torch.log(d) + b
print(loss, d.item(), b)
print(iou)
```
